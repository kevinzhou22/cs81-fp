```python
import pickle
import os
from matplotlib import pyplot as plt
```

#### Training the DQN

```python
path = os.getcwd()
file = '\dqn_exp2.pkl'
with open(path + file, 'rb') as output:
    a = pickle.load(output)
```

```python
plt.title("Sum of rewards across training episodes")
plt.plot(a)
plt.xlabel("Training episodes")
plt.ylabel("Sum of rewards")
plt.show()
```

#### Remarks
1. Here, our desired distance of following is d = [0.5, 3], q = 0.40.
2. The sum of rewards across the training episodes thus far tells us that we haven't converged to a good DQN yet. As pointed out in a previous remark, a negative net reward suggests that we are not following the target as desired. But this outcome is prevalent especially in the first 30 training episodes where we are still exploring via taking random actions. But as we start exploiting the prediction of the DQN after this checkpoint, we are already getting some positive net rewards in some episodes. This result is not stable, however, as the net rewards dips to negative once again. Although, we can notice a slight upward trend of the net rewards. Further, note that our exploitation is still warming up as we are still taking some random actions, and so deviations from the desired distance can still be observed. In other words, the DQN that we have constructed is still in the process of converging to better model weights to come up with better policy.
